```python
import pymatgen
from pymatgen.ext.matproj import MPRester
m = MPRester('J9RM5ogTNGLHFamO')
q= m.query({'e_above_hull':{'$lt':0.1},'band_gap':{'$gt':0.1}},['material_id','piezo','spacegroup','structure','dos'], chunk_size=500, max_tries_per_chunk=5, mp_decode=True)
```

```python
X=[]
Y=[]
for i in range(len(q)):
    sn=q[i]['spacegroup']['number']
    if (sn==1 or 3<=sn<=9 or 16<=sn<=46 or 75<=sn<=82 or 89<=sn<=122 or 143<=sn<=146 or 149<=sn<=161 or 168<=sn<=174 or 177<=sn<=190 or 195<=sn<=199 or 207<=sn<=220):
        if q[i]['piezo']==None:
            Y.append([q[i]['material_id'],q[i]['structure'],q[i]['dos'],q[i]['piezo']])
        else:
            X.append([q[i]['material_id'],q[i]['structure'],q[i]['dos'],q[i]['piezo']['eij_max']])
```

```python
import pandas as pd
df = pd.DataFrame(X, columns=['material_id','structure','dos','piezo'])
```
