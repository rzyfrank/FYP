```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('strcom_cfid_dropna.csv')
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
y = df.iloc[:,1923].values
X = df.iloc[:,0:1923].values
```

```python
y.shape, X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(50,20), activation='relu', solver='sgd', alpha=0.0001, batch_size='auto', learning_rate ='constant')
```

```python
A = [(i*100, j*100, k*10) for i in range(5,9) for j in range(1,5) for k in range(1,10)]
```

```python
from sklearn.model_selection import GridSearchCV

para_test = {'hidden_layer_sizes' : A}
gsearch = GridSearchCV(estimator= MLPClassifier(activation='relu', solver='sgd', alpha=0.0001, batch_size='auto', learning_rate ='constant'), param_grid= para_test, cv= 5, scoring='roc_auc')
gsearch.fit(X,y)
gsearch.best_params_, gsearch.best_score_
```

```python
mlp.fit(X_train,y_train)
```

```python
from sklearn import metrics

y_pred = mlp.predict(X_test)
y_predprob = mlp.predict_proba(X_test)[:,1]

print("Accuracy : %.4g" % metrics.accuracy_score(y_test, y_pred))
print("AUC Score (Train): %f" % metrics.roc_auc_score(y_test, y_predprob))
```
